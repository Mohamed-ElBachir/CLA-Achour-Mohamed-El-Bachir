```python
def beginWithAnA(name):
    return name[0]=="A"
```

```python
def namesThatBeginsWithAnA(students_name):
    for name in students_name:
        if(beginWithAnA(name)): print(name)
```

```python
file = open("student_names.txt")
names = file.read()
names = names.split()
namesThatBeginsWithAnA(names)
```

```python
def check_if_palindrome(string):
    left_position = 0
    right_position = len(string) - 1
    
    while right_position >= left_position:
        if not string[left_position] == string[right_position]:
            return False
        left_position += 1
        right_position -= 1
        return True
```

```python
# Let's check if our function works:
check_if_palindrome(('GDG'))
```

```python
# Let's have another example :
check_if_palindrome("CLA")
```

```python
#A prime number is a number that is divisible only by itself and 1 
def prime_func(n):
    """
    This function take a parameter n which is a natural number bigger than 0 
    return : a boolean value that tell us if the number is prime or not 
    """
    if (n==1):
        return False
    elif (n==2):
        return True;
    else:
        for x in range(2,n):
            if(n % x==0):
                return False
        return True             
```

```python
# Let's check if our function works
prime_func(7)
```

```python
# Write a Python function to calculate the factorial of a number
def factorial(n):
    """This function return the factorial of a given number (the num must be a positive integer)
        for example 5! = 5 * 4 * 3 * 2 * 1 = 120
        it returns the value of the given number 
    """
    if n <0:
        print("Please give a positive integer ")
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(-5)
```

```python
factorial(6)
```

```python
# reverse the string
def reverse(string):
    """For a given string string this function print the string from the end the beginning
    example : "abcde" gonna be "edcba"
    """
    string = string[::-1]
    return string
```

```python
reverse("abcdefg")
```

```python
# Sum of a list 
def sum(numbers):
    total = 0
    for n in numbers:
        total += n
    return total
```

```python
sum([1,2,3,4,5,6,7,8])
```

```python
# Write a Python function to find the Max of three numbers.

def maximum(a, b, c):
  
    if (a >= b) and (a >= c):
        largest = a
  
    elif (b >= a) and (b >= c):
        largest = b
    else:
        largest = c
          
    return largest
```

```python
maximum(1,19,9)
```
