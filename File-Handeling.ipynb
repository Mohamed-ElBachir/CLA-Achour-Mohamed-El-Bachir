```python
# Read all of the content of the file in one variable.

def file_read(fname):
    f_name = fname
    txt = open(fname)
    print(txt.read())

file_read('student_names.txt')
```

```python
# Read the first n lines of the file.

def read_from_head(fname , nlines):
        from itertools import islice
        with open(fname) as f:
                for line in islice(f, nlines):
                        print(line)
read_from_head('student_names.txt',1)
```

```python
def LastNlines(fname , N):
    with open(fname) as file: 
        # loop to read iterate
        # last n lines and print it
        for line in (file.readlines() [-N:]):
            print(line, end ='')
```

```python
LastNlines(fname="student_names.txt" , N = 1)
```
